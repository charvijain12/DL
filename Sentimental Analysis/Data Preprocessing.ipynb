# 4. Combine datasets
<p> convert dataset 2 content into dataset 1 numbers as per emotion mentioned above</p>

```python
# before changes are made
df2.info()
```

```python
# drop tweet_id column
df2.drop(['tweet_id'], axis = 1, inplace = True)

# replace content column with numeric values mapped to dataset 1
df2['sentiment'].replace(['sadness', 'boredom', 'surprise'], 0, inplace = True)
df2['sentiment'].replace(['enthusiasm', 'happiness', 'fun', 'neutral'], 1, inplace = True)
df2['sentiment'].replace(['love'], 2, inplace = True)
df2['sentiment'].replace(['hate', 'anger', 'empty', 'relief'], 3, inplace = True)
df2['sentiment'].replace(['worry'], 4, inplace = True)
```

```python
# after changes are made
df2.info()
```

# 5. Analysis for Cleaning the dataset 
1. Need to remove unnecessary words from 'text' column

```python
# if using new notebook for a saved dataset run this cell
df = pd.read_csv("./DL_Dataset.csv") # change name as per whatever you have saved it as
df.head()
```

```python
# drop Unnamed: 0 column 
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.head()
```

```python
len(df)
```

```python
first_word = list()
for i in range(len(df)):
    word = df['text'][i].split()[0]
    first_word.append(word)

first_word = pd.Series(first_word)
unique_first_word = first_word.unique()
```

```python
# view result
print(len(unique_first_word))
for w in unique_first_word:
    print(w)
```

<b> too long to check.</b>
<p> What we need is starting words that are not alphabets</p>

```python
print("ASCII for 'a': {} \nASCII for 'z': {} \nASCII for 'A': {} \nASCII for 'Z': {}". format(ord('a'), ord('z'), ord('A'), ord('Z')))
print("\n\nASCII for 0: {} \nASCII for 9: {}". format(ord('0'), ord('9')))
```

```python
non_alphabet_words = list()
for w in unique_first_word:
    ascii_val = ord(w[0])
    if (ascii_val < 65) or (ascii_val > 122) or (ascii_val > 90 and ascii_val < 97):
        non_alphabet_words.append(w)
```

```python
# view result
print(len(non_alphabet_words))
for w in non_alphabet_words:
    print(w)
```

# 6.1 Repeat the below steps till the cell below shows 0

```python
row_index = list() # list of rows that have no value after removing unwanted words
not_ok_count = 0
for i in range(len(df)):
    start = ord(df['text'][i][0])
    sentence = df['text'][i]
    if (start >= 48 and start <= 57) or (start >= 97 and start <= 122) or (start >= 65 and start <= 90):
        continue
    else:
        not_ok_count += 1
        print(i, end= "\t")
        try:
            first_space = 0
            if first_space == 0:
                first_space = df['text'][i].index(" ")

            print(first_space)
            new_sentence = sentence[first_space + 1:]
            print(sentence, new_sentence, sep = "\t")
            df['text'][i] = new_sentence
        except:
            print("NEED TO REMOVE ROW ", i)
            row_index.append(i)

print(not_ok_count)
        
```

```python
# list of rows to remove
print(row_index)
```

```python
df.drop(index=row_index, axis = 0, inplace = True)
df = df.reset_index()
df.drop(['index'], axis = 1, inplace = True)
df.info()
```
